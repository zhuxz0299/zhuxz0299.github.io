```python
import numpy as np
import matplotlib.pyplot as plt

# 定义第一个方程的函数
def equation_1(u):
    return -u / np.tan(u)

# 定义第二个方程的函数
def equation_2(u, m, U_0, h_bar, a):
    return np.sqrt((2 * m * U_0) / (h_bar**2) * a**2 - u**2)

def equation_3(u):
    return np.sqrt(5 - u**2)

# 设置参数值
m = 1.0
U_0 = 1.0
h_bar = 1.0
a = 1.0

# 生成一组u值
u_values = np.linspace(0, 10, 1000)

# 计算对应的v值
v_values = equation_2(u_values, m, U_0, h_bar, a)

# 创建图形
fig, ax = plt.subplots()

# 绘制第一个方程的图像
ax.plot(u_values, equation_1(u_values))

# 绘制第二个方程的图像
ax.plot(u_values, v_values)

# 绘制第三个方程的图像
ax.plot(u_values, equation_3(u_values))

# 设置坐标轴位置和方向
ax.spines['left'].set_position('zero')
ax.spines['left'].set_color('gray')
ax.spines['bottom'].set_position('zero')
ax.spines['bottom'].set_color('gray')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# 调整坐标轴标签位置
ax.xaxis.set_label_coords(1, 0.5)
ax.yaxis.set_label_coords(0.5, 1)

# 添加图例
ax.legend()

plt.ylim(0, 10)
# plt.axis('equal')

# 显示图形
plt.show()
```
