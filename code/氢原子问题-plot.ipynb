```python
import numpy as np
import matplotlib.pyplot as plt

def W10(a, r):
    return (2 / a**(3/2) * np.exp(-r / a))**2 * r**2

def W20(a, r):
    return (1 / np.sqrt(2) / a**(3/2) * (1 - r / (2 * a)) * np.exp(-r / (2 * a)))**2 * r**2

def W21(a,r):
    return (1 / (2 * np.sqrt(6) * a**(3/2)) * (r / a) * np.exp(-r / (2 * a)))**2 * r**2

def W30(a, r):
    return (2 / (3 * np.sqrt(3) * a**(3/2)) * (1 - (2 * r) / (3 * a) + (2 / 27) * (r / a)**2) * np.exp(-r / (3 * a)))**2 * r**2

def W31(a, r):
    return (8 / (27 * np.sqrt(6) * a**(3/2)) * (1 - r / (6 * a)) * np.exp(-r / (3 * a)) * (r / a))**2 * r**2

def W32(a, r):
    return (4 / (81 * np.sqrt(30) * a**(3/2)) * (r / a)**2 * np.exp(-r / (3 * a)))**2 * r**2

# 设置参数
a = 1.0  # a 的值
r = np.linspace(0, 15, 100)  # r 的范围

# 计算函数值
W10_values = W10(a, r)
W20_values = W20(a, r)
W21_values = W21(a, r)
W30_values = W30(a, r)
W31_values = W31(a, r)
W32_values = W32(a, r)

# 绘制图像
fig, axs = plt.subplots(2, 3, figsize=(15, 10))

axs[0,0].plot(r, W10_values, label='W10')
axs[0,0].set_xlabel('r/a')
axs[0,0].legend()

axs[1,0].plot(r, W20_values, label='W20')
axs[1,0].set_xlabel('r/a')
axs[1,0].legend()

axs[0,1].plot(r, W21_values, label='W21')
axs[0,1].set_xlabel('r/a')
axs[0,1].legend()

axs[1,1].plot(r, W30_values, label='W30')
axs[1,1].set_xlabel('r/a')
axs[1,1].legend()

axs[0,2].plot(r, W31_values, label='W31')
axs[0,2].set_xlabel('r/a')
axs[0,2].legend()

axs[1,2].plot(r, W32_values, label='W32')
axs[1,2].set_xlabel('r/a')
axs[1,2].legend()

plt.show()
```
